Use the dataset of Game, we have explored earlier and perform following operations.

* Calculate the total global sales for all games in the dataset.

* Find the top 5 most popular game genres based on global sales.
* Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales).
* Display last 10 rows.
* Determine the top 3 gaming platforms with the highest global sales.
* Calculate the market share (percentage of global sales) for each genre.
* Calculate the correlation coefficient between NA_Sales and EU_Sales. (.corr function)

```python
import pandas as pd

data=pd.read_csv("vgsalesGlobale.csv")
data.head()
#------------------------------ total global sale -----------------------
total=data['Global_Sales'].sum()
print("Total Global Sale, ",total)


#------------------------------ Sale for each region -----------------------
s1=data['NA_Sales'].mean()
s2=data['EU_Sales'].mean()
s3=data['JP_Sales'].mean()
s4=data['Other_Sales'].mean()

print(f'\n\nAverage Sale for EU is {s2}, JP is {s3}, NA is {s1} and others {s4} ')

#---------------------------------- Display last 10 rows.----------------------
data.tail(10)
```

```python
#------------------------------ highest global sale -----------------------
highest=data.sort_values('Global_Sales',ascending=False)
print("\n\n Top 5 highest selling games are: ")
highest[['Name','Global_Sales']].head()
```

```python
#------------------------------ Top 5 genres -----------------------
highest=data.sort_values('Global_Sales',ascending=False)
print("\n\n Top 5 highest selling games are: ")
highest[['Genre','Global_Sales']].head()
```

```python
#-----------------------------Calculate the total sales for each genre--------------------
na=data.groupby('Genre')[['NA_Sales','Global_Sales']].sum()
eu=data.groupby('Genre')[['EU_Sales','Global_Sales']].sum()
jp=data.groupby('Genre')[['JP_Sales','Global_Sales']].sum()
ot=data.groupby('Genre')[['Other_Sales','Global_Sales']].sum()
print('sales for each genre in each region\n\n',na,eu,jp,ot)
```

```python
#-----------------------------Platforms--------------------
highest=data.sort_values('Global_Sales',ascending=False)
print("\n\n Top 3 gaming platforms  are: ")
highest[['Platform','Global_Sales']].head(3)
```

```python
#-----------------------------Market share--------------------
sale=data.groupby('Genre')['Global_Sales'].sum()
gsale=sale.sum()
print('Market share is: ',sale/gsale*100)
```

```python
#-------------------------------Corelation--------------------
co=data['NA_Sales'].corr(data['EU_Sales'])
print('Correlation bw NA and EU is: ',co)
```


## Task 3 Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
def classify(emails):

    important = []
    promotions = []
    spam = []
    imp = ['urgent', 'important', 'action required', 'please respond']
    prom = ['sale', 'discount', 'offer', 'limited time', 'buy now']
    sp = ['win', 'free', 'click here', 'congratulations', 'prize', 'exclusive offer']

    for email in emails:

        if any(k in email for k in imp):
            important.append(email)

        elif any(k in email for k in prom):
            promotions.append(email)

        elif any(k in email for k in sp):
            spam.append(email)

    return imp, prom, sp

emails = [
    "URGENT: Action Required for Your Account",
    "Limited Time Offer: 50% off your next purchase",
    "Congratulations! You have won a free prize",
    "Please respond: Important Update for Your Account",
    "Exclusive Offer: Buy 1 get 1 free on all items",
    "Click here to claim your prize"
]

important, promotions, spam = classify(emails)

print("Important Emails:")
for email in important:
    print(email)

print("\nPromotions Emails:")
for email in promotions:
    print(email)

print("\nSpam Emails:")
for email in spam:
    print(email)
```

## Task 4 Online Store Shopping Cart
Write a Python program to simulate the online store's shopping cart system. The program should use a while loop to provide an interactive shopping experience.


### Tasks:
* Initialize an empty list called shopping_cart to store items.
* Display a welcome message and available products.
* Loop should be iterating over AI Response Simulator and Investment decision and keep asking for user input.
* Prompt the user to enter the number corresponding to the product they want to add to the cart.
* Check if the user entered 'q'. If so, break the loop.
* Convert the choice to an integer and validate it within the valid range (1 to 3).
* Based on the user's choice, add the selected product to the shopping_cart list.
* Display a confirmation message indicating the added product.
* Continue the loop to allow the user to add more products or proceed to checkout.
* After the loop ends, display the contents of the shopping cart.
* Display a goodbye message.
* End the program.

```python
shopping_cart = []
print('---------------------------Welcome to Shop-----------------\n\n\nHere are the Products\n\n')

prod = {
    "1": "T-shirt 300pkr",
    "2": "Sweat Shirt 400pkr",
    "3": "Hoodie 500pkr"
}

for x in prod:
    print(x, prod[x])

choice = ''
while choice != 'q':
    ask = input('Select From the menu ( 1, 2, 3 ) or press "q" to exit and checkout: ')
    if ask == 'q':
        choice = ask
    elif ask in prod:
        shopping_cart.append(prod[ask])
        print('Added into your cart.\n')
    else:
        print('Incorrect choice.\n')

print('\nYour Shopping Cart:')
for x in shopping_cart:
    print(x,'\n\nGOOD BYE :)')
```

## Task 6

```python
import numpy as np
arr1=np.array([10,23,32,45,67,44,34,54,21,1,2,3])
arr2=np.array([20,33,52,15,64,42,31,34,13,19,20,29])
x=arr1.reshape(3,4)
y=arr2.reshape(3,4)
print(f'Arr1:\n{x}\n\nArr2:\n{y}')

z=np.concatenate((x,y))
print(f'\n\nConcatinated array:\n{z}')
```
